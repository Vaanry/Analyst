### Задание 3

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

#### Событие № 1

*Для начала необходимо определиться с длительностью игрового ивента. Ивенты могут длиться несколько часов, дни, недели, месяц.
Очётный период нужно определять, исходя из длительности. Предположим, что данный игровой ивент длится неделю (например, приурочен в каким-то праздникам). Тогда отчётным периодом также будет неделя - всё то время, пока шёл ивент.*

Поделим метрики на 2 группы: экономические и игровые. Экономические метрики будут показывать нам прибыльность (или убыточность) данного ивента. Игровые покажут влияние игровых механик на поведение игроков.

##### Экономические метрики:

Для существующих игроков.

1. Прежде всего оценим вовлечённость игроков в событие. Для этого посчитаем соотношение кол-ва игроков, принявших участие в ивенте (Event Users, EU), к общей недельной аудитории. Назовём эту метрику Event Engagement - EE. 
EE = EU / WAU 

2. Далее необходимо оценить, как изменится Sticky Factor - показатель увлекательности игры для юзера. Будем считать так:
Sticky Factor = WAU / MAU * 100%. Посчитаем Sticky Factor для недели с ивентом, а также высчитаем средний недельный Sticky Factor за месяц перед событием. После этого оценим изменение данного показателя во время проведения события относительно усредненного Sticky Factor за предыдущий месяц.

Sticky Factor Event = WAU Event / MAU * 100%

Mean Sticky Factor = sum(WAU n / MAU * 100%) / N, где n - номер недели, а N - общее кол-во недель перед событием.

Sticky Factor Diff = (Sticky Factor Event - Mean Sticky Factor) / Mean Sticky Factor - получаем изменение Sticky Factor во время проведения события относительно предыдущего периода.

3. Sessions (время, которое пользователь провел в мобильном приложении) + Average Session Duration (среднее время сессий за период времени). По аналогии с предыдущим показателем, будем оценивать, как изменятся эти метрики во время игрового события по сравнению с предыдущим периодом.

4. Retention. Пожалуй, это самый важный показатель, т.к. сравнивая Retention во время проведения ивента и в предыдущий период, мы можем оценить, насколько данное событие способствует возврату игроков.

5. ARPWAU - с помощью этой метрики мы узнаем, как влияет ивент на монетизацию нашей игры.

Для новых игроков.

1. Привлечение (Acquisition). Посчитаем, изменился ли приток новых игроков во время ивента.
Acquisition = New Users / Total Users.

2. CR - увеличится ли конверсия в покупку игры во время проведения игрового события?

##### Игровые метрики:

1. Success rate (SR) - отношение количества игроков, успешно завершивших событие, к общему количеству игроков, принявших участие.

2. Levels Rating (LR) - посмотрим, как игроки (не дошедшие до финала) распределяются по уровням.

3. Time Event Rate (TER) - соотношение времени, потраченного на прохождение ивента, к Average Session Duration. Если этот показатель высокий, то значит игроки заходили в игру только ради события, что является хорошей характеристикой нашего ивента.

На основании этих метрик мы можем сделать вывод о том, насколько сбалансирована механика данного ивента. Если слишком много пользователей быстро доходят до финала, то при высоком TER это может означать, что они быстро пройдут все уровни события и покинут игру. Если, напротив, слишком мало пользователей дойдут до финала, то это означает, что либо событие неинтересное само по себе, либо слишком сложное, так что игроки попросту теряют интерес. Дополнительную информацию даст LR - в идеале это должно быть несколько скошенное вправо (в сторону финала) распределение. Полученные результаты мы передадим разработчикам, чтобы они дальше продумывали игровой баланс.

#### Событие № 2

Прежде всего, стоит оставить все метрики, перечисленные выше, и после проведения обоих событий сравнить ключевые показатели (например, Retention и ARPWAU). На основании этого сравнить, какая механика является более экономически успешной, т.к. в конечном итоге нас интересует именно прибыльность игры.
Важными показателями будут TER и Sticky Factor, т.к. они покажут, насколько усложнение механики подогреет интерес игроков. SR и LR будут менее показательными, т.к. при усложнении механики распределение ожидаемо сдвинется влево (а если не сдвинется - есть повод задать вопрос разработчикам, а не схалтурили ли они при разработке новых механик?)
Дополнительно стоит ввести следующие метрики:

1. Churn Rate. Посмотрим, насколько усложнение повлияет на отток игроков. В идеале стоит смотреть этот показатель по всем уровням. И если на каком-то игровом уровне он повышен - стоит указать на этот уровень разработчикам. Возможно, его стоит упростить.

2. Try Rate (TR) - среднее количество попыток прохождения ивента после каждого отката. Отдельно стоит посмотреть, есть ли корреляция между TR и Churn Rate. Возможно, при слишком высокой сложности игроки будут бросать игру после какой-то попытки. Тогда стоит упростить механику. Если этот показатель будет сильно коррелировать с Churn Rate по всем уровням - то, возможно, данную механику (откат на несколько уровней) вообще стоит убрать.
